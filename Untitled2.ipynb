```python
import tgt

tg = tgt.io.read_textgrid('step1.TextGrid.utf8')
ophonetier = tg.get_tier_by_name('IU/phone')
owordtier =  tg.get_tier_by_name('Word')

phonetier = ophonetier.get_copy_with_gaps_filled()
wordtier  = owordtier.get_copy_with_gaps_filled()

pst = phonetier.start_time
pet = phonetier.end_time

wst = wordtier.start_time
wet = wordtier.end_time

st = pst
et = pet
```

```python
print(phonetier)
print(wordtier)
```

```python
timeptr = st
ann = wordtier.get_annotation_by_start_time(st)
wtimelist = []
while timeptr < et:
    if ann is not None:
        wtimelist.append(ann.start_time)
        timeptr = timeptr + ann.duration()
        ann = wordtier.get_annotation_by_start_time(timeptr)
    else:
        break

print(wtimelist)
```

```python
timeptr = st
ann = phonetier.get_annotation_by_start_time(st)
ptimelist = []
while timeptr < et:
    if ann is not None:
        ptimelist.append(ann.start_time)
        timeptr = timeptr + ann.duration()
        ann = phonetier.get_annotation_by_start_time(timeptr)
    else:
        break

print(ptimelist)
```

```python
idx = 0
for p in ptimelist:
    if (p != wtimelist[idx]):
        print("Not equal")
        print("Absolute: :", abs(p - wtimelist[idx]))
        if (abs(p - wtimelist[idx]) < 0.01):
            print("adjusting")
    if not(p >= wtimelist[idx] and p < wtimelist[idx+1]):
        idx = idx + 1

    #print('{0:.3g}'.format(p), "[", '{0:.3g}'.format(wtimelist[idx]), '{0:.3g}'.format(wtimelist[idx+1]), "]")
    print(p, "[", wtimelist[idx], wtimelist[idx+1], "]", abs(p - wtimelist[idx]))
```
